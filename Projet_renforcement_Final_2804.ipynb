</h1>

La m√©thodologie d'alimentation des donn√©es d√©bute par le chargement de deux DataFrames √† partir des fichiers CSV, "API_Dataset.csv" contenant des donn√©es relatives √† un utilisateur, et "API_df_account.csv" repr√©sentant une liste de noms d'utilisateurs √† suivre. Ensuite, deux fonctions sont d√©finies : <b>get_instagram_data()</b> pour extraire les informations d'un compte Instagram, et <b>get_followings()</b> pour obtenir les noms d'utilisateur des comptes suivis par un utilisateur donn√©.

Par la suite, une it√©ration est effectu√©e sur les comptes suivis r√©pertori√©s dans le DataFrame df_account. Pour chaque compte, les donn√©es Instagram sont r√©cup√©r√©es √† l'aide de la fonction <b>get_instagram_data()</b>, puis ajout√©es √† un DataFrame combin√©. Un contr√¥le est √©galement effectu√© pour v√©rifier si le nombre de lignes dans API_df_account est inf√©rieur √† 1000. Dans ce cas, les comptes suivis par le compte actuel sont r√©cup√©r√©s √† l'aide de <b>get_followings()</b> et ajout√©s √† df_account.

Une fois les donn√©es collect√©es pour chaque compte, elles sont int√©gr√©es au DataFrame principal df. Les DataFrames mis √† jour sont ensuite sauvegard√©s dans des fichiers CSV. Tout au long du processus, un d√©lai de 20 secondes est respect√© entre chaque it√©ration pour √©viter tout probl√®me de blocage de compte ou de limitation de taux par Instagram. Enfin, une fois toutes les it√©rations termin√©es, le script se ferme en fermant la session Instaloader.

```python

#Charger les DataFrames
df = pd.read_csv("API_Dataset.csv", sep=",") #Dataset df
df_account = pd.read_csv("API_df_account.csv", sep=",") #Liste des comptes suivis du compte d'origine (xenon_ne)

df = df.drop_duplicates()

#Fonction pour r√©cup√©rer les informations d'un compte Instagram
def get_instagram_data(loader,username):
    
    #R√©cup√©rer le profil de l'utilisateur
    profile = Profile.from_username(loader.context, username)
    #R√©cup√©rer les publications du profil
    posts = list(profile.get_posts())

    #Initialiser les variables pour les totaux
    total_likes = 0
    total_comments = 0
    total_video_posts = 0
    all_hashtags = []

    #Boucle √† travers les publications pour obtenir les informations n√©cessaires
    for post in tqdm(posts, desc="Processing Posts", unit="post"):
        total_likes += post.likes
        total_comments += post.comments
        total_video_posts += post.is_video
        all_hashtags.extend(post.caption_hashtags)

    #Compter les occurrences de chaque hashtag
    hashtags_count = Counter(all_hashtags)

    #R√©cup√©rer les 10 hashtags les plus courants
    top_categories = hashtags_count.most_common(10)
    top_category_list = [category for category, _ in top_categories]

    #Retourner les donn√©es du profil
    return {
        'nom_utilisateur': profile.username,
        'nom_complet': profile.full_name,
        'biographie': profile.biography,
        'external_url': profile.external_url,
        'nb_followers': profile.followers,
        'nb_suivis': profile.followees,
        'id_compte': profile.userid,
        'compte_prive': profile.is_private,
        'compte_verifie': profile.is_verified,
        'nb_post': profile.mediacount,
        'nb_video_posts': total_video_posts,
        'url_image_profil': profile.profile_pic_url,
        'nb_likes_total': total_likes,
        'nb_comments_total': total_comments,
        'category_hashtag': ", ".join(top_category_list)
    }

def get_followings(loader,username):
    profile = Profile.from_username(loader.context, username)
    followings = []
    for following in profile.get_followees():
        followings.append(following.username)
    return followings

i = 0

#It√©ration sur les lignes du DataFrame df_account
for index, row in tqdm(df_account.iterrows(),desc='It√©ration des users'):
    time.sleep(20)
    #Initialiser Instaloader
    loader = Instaloader()
    loader.login("********", "********")
    
    if index >= 1000:
        break
    nom = row["Utilisateur Suivi"]
    print("\n")
    print(f"-----------{nom}--------------")
    print("\n")
    df_combined = pd.DataFrame()
    
    user_data = get_instagram_data(loader, nom)
    
    # k=0
    # while True:
    #     k+=1
    #     if k ==3:
    #         break
    #     try: 
    #         user_data = get_instagram_data(loader, nom)
    #         break  # Sort de la boucle si aucune exception n'est lev√©e
    #     except Exception as e:
    #         print(f"Erreur lors de la r√©cup√©ration des donn√©es Instagram : {e}")
    #         print("Nouvelle tentative dans 20 secondes...")
    #         # Initialiser Instaloader
    #         loader = Instaloader()
    #         loader.login("rem_wf202", "dadou2000")
    #         time.sleep(20)
    
    # if k == 3:
    #     continue
    
    df_combined = pd.concat([df_combined, pd.DataFrame([user_data])], ignore_index=True)
    
    # Suppression du compte trait√©
    df_account.drop(index, inplace=True)
    
    df_account1 = df_account
    
    if len(df_account) < 1000:
        followings = get_followings(loader,nom)  # Cette fonction doit √™tre d√©finie dans votre code
        followings_df = pd.DataFrame({"Utilisateur Suivi": followings})
        df_account = pd.concat([df_account, followings_df], ignore_index=True)
    
    df = pd.concat([df, df_combined], ignore_index=True)
    
    df.to_csv("API_Dataset.csv",index=False)
    i+=1
    
    df_account1.to_csv("API_df_account.csv",index=False)
    print("\n")
    print(f"{i} : {nom} ajout√©")
    loader.close()  
```


<h1 style="background-color: #e57373; padding: 10px; width: 600px; font-size: 20px; text-align: center; box-shadow: 0 4px 8px 0 rgba(0, 0, 0, 0.15), 0 6px 20px 0 rgba(0, 0, 0, 0.15);">Pourquoi cela ne peut-il pas √™tre utilis√© dans notre cas ?</h1>


Cette m√©thode collecte une grande quantit√© de donn√©es sur un utilisateur. Cependant, nous n'avons aucun moyen de r√©cup√©rer les pr√©f√©rences des utilisateurs ni les utilisateurs qui ont aim√© un post (voir ci-dessous le code). <center><h1>Algorithme de Recommandation</h1></center>
<hr style="border-width:6px;border-color:#75DFC1">

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    L'algorithme de recommandation employ√© dans ce projet est bas√© sur le Q-Learning, une m√©thode d'apprentissage par renforcement. Initialement, la table Q est remplie de valeurs nulles, symbolisant les estimations de d√©part pour la qualit√© des actions. Le syst√®me parcourt divers √©pisodes, ajustant ses recommandations en fonction des interactions et des r√©compenses re√ßues, pour am√©liorer progressivement sa capacit√© √† pr√©dire les pr√©f√©rences des utilisateurs.
</p>

<h2 style="font-family: 'Arial', sans-serif; font-size: 18px; color:#75DFC1;">M√©triques d'√âvaluation</h2>
<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Les performances du syst√®me de recommandation sont √©valu√©es √† l'aide des m√©triques suivantes :
</p>

<ul style="font-family: 'Arial', sans-serif; font-size: 14px;">
    <li><strong>Taux de Pertinence :</strong> Ce taux mesure la proportion de recommandations qui correspondent effectivement aux pr√©f√©rences de l'utilisateur. Il est calcul√© par la formule suivante : <br>
        <b>Taux de pertinence = (Nombre de posts recommand√©s pertinents / Nombre total de posts recommand√©s) * 100</b>
    </li>
    <li><strong>Taux de Couverture :</strong> Cette m√©trique √©value dans quelle mesure les recommandations couvrent les diff√©rentes pr√©f√©rences de l'utilisateur. Sa formule est : <br>
        <b>Taux de couverture = (Nombre de pr√©f√©rences de l'utilisateur couvertes / Nombre total de pr√©f√©rences de l'utilisateur) * 100</b>
    </li>
    <li><strong>Diversit√© des Recommandations :</strong> Indique le nombre de cat√©gories uniques parmi les posts recommand√©s, promouvant une vari√©t√© dans les suggestions faites √† l'utilisateur.</li>
    <li><strong>Popularit√© Moyenne des Posts Recommand√©s :</strong> Repr√©sente la moyenne des likes sur les posts recommand√©s, donnant une id√©e de la popularit√© des suggestions au sein de la communaut√© des utilisateurs.</li>
    <li><strong>Nouveaut√© des Recommandations :</strong> Mesure la proportion de posts non encore explor√©s par l'utilisateur, enrichissant son exp√©rience par la d√©couverte de nouveaux contenus.</li>
</ul>


```python
#Choisir une action (recommander un post) en utilisant Q-Learning
def choose_action(state, q_table, exploration_rate):
    """
    Choisit une action (recommande un post) en utilisant l'algorithme Q-Learning.

    Parameters:
    state (int): L'√©tat actuel du syst√®me, qui peut repr√©senter l'utilisateur ou une autre caract√©ristique de l'environnement.
    q_table (numpy.array): La table Q, qui contient les valeurs Q pour chaque paire (√©tat, action).
    exploration_rate (float): Le taux d'exploration, qui d√©termine la probabilit√© d'explorer une nouvelle action au lieu de choisir celle avec la valeur Q maximale.

    Returns:
    int: L'indice de l'action choisie, correspondant √† l'indice de la colonne dans la table Q.
    """
    if np.random.uniform(0, 1) < exploration_rate:
        return np.random.choice(q_table.shape[1])
    else:
        return np.argmax(q_table[state, :])


#Mettre √† jour la table Q
def update_q_table(q_table, state, action, reward, next_state, learning_rate, discount_factor):
    """
    Met √† jour la table Q en utilisant l'algorithme Q-Learning.

    Parameters:
    q_table (numpy.array): La table Q, qui contient les valeurs Q pour chaque paire (√©tat, action).
    state (int): L'√©tat actuel du syst√®me.
    action (int): L'action effectu√©e dans l'√©tat actuel.
    reward (float): La r√©compense re√ßue apr√®s avoir effectu√© l'action dans l'√©tat actuel.
    next_state (int): L'√©tat suivant du syst√®me apr√®s avoir effectu√© l'action.
    learning_rate (float): Taux d'apprentissage qui contr√¥le dans quelle mesure les nouvelles informations sont int√©gr√©es √† la table Q.
    discount_factor (float): Facteur de r√©duction qui contr√¥le l'importance des r√©compenses futures par rapport aux r√©compenses imm√©diates.
    """
    best_next_action = np.argmax(q_table[next_state, :])
    td_target = reward + discount_factor * q_table[next_state, best_next_action]
    td_error = td_target - q_table[state, action]
    q_table[state, action] += learning_rate * td_error


#Entra√Ænement du mod√®le Q-Learning
def train_q_learning(df, learning_rate=0.1, discount_factor=0.9, exploration_rate=0.1, num_episodes=1000):
    """
    Entra√Æne le mod√®le Q-Learning.

    Parameters:
    df (pandas.DataFrame): Le DataFrame contenant les donn√©es d'interaction utilisateur-post.
    learning_rate (float, optionnel): Le taux d'apprentissage pour la mise √† jour des valeurs Q. Par d√©faut, 0.1.
    discount_factor (float, optionnel): Le facteur d'escompte pour les r√©compenses futures. Par d√©faut, 0.9.
    exploration_rate (float, optionnel): Le taux d'exploration pour choisir une action. Par d√©faut, 0.1.
    num_episodes (int, optionnel): Le nombre d'√©pisodes d'entra√Ænement. Par d√©faut, 1000.
    """

    #Nombre d'utilisateurs et de posts uniques
    num_users = df['user_id'].nunique()
    num_posts = df['post_id'].nunique()

    #Espace des actions et des √©tats
    action_space = num_posts
    state_space = num_users

    #Initialisation de la table Q avec des z√©ros
    q_table = np.zeros((state_space, action_space))

    #Boucle sur le nombre d'√©pisodes
    for _ in range(num_episodes):
        #S√©lection al√©atoire d'un utilisateur
        user_id = np.random.randint(0, num_users)
        #Extraction des interactions de l'utilisateur s√©lectionn√©
        user_df = df[df['user_id'] == user_id]
        #D√©finition de l'√©tat initial
        state = user_id
        #Choix d'une action selon la politique d'exploration
        action = choose_action(state, q_table, exploration_rate)
        #R√©cup√©ration de la r√©compense associ√©e √† l'action choisie
        reward = user_df.iloc[action]['Likes']
        #L'√©tat suivant est identique √† l'√©tat actuel dans ce cas
        next_state = user_id
        #Mise √† jour de la table Q
        update_q_table(q_table, state, action, reward, next_state, learning_rate, discount_factor)

    return q_table


# Recommender system
def recommend_posts(user_id, q_table, num_recommendations=10):
    """
    Recommande des posts √† un utilisateur en utilisant la table Q.

    Parameters:
    user_id (int): L'identifiant de l'utilisateur pour lequel recommander des posts.
    q_table (numpy.array): La table Q contenant les valeurs Q pour chaque paire (√©tat, action).
    num_recommendations (int, optionnel): Le nombre de posts √† recommander. Par d√©faut, 10.

    Returns:
    tuple: Un tuple contenant une liste des identifiants des posts recommand√©s et les pr√©f√©rences de l'utilisateur.
    """
    #S√©lection des interactions de l'utilisateur sp√©cifi√©
    user_df = df[df['user_id'] == user_id]
    #Extraction des pr√©f√©rences de l'utilisateur
    user_preferences = user_df[['user_preference_1', 'user_preference_2', 'user_preference_3', 'user_preference_4', 'user_preference_5']].iloc[0]

    recommended_posts = []
    for _ in range(num_recommendations):
        #S√©lection de l'action (post) avec la plus grande valeur Q pour l'utilisateur donn√©
        action = np.argmax(q_table[user_id, :])
        #Ajout du post recommand√© √† la liste des recommandations
        recommended_posts.append(action)
        #Marquer le post comme d√©j√† recommand√© en affectant une valeur n√©gative √† sa valeur Q
        q_table[user_id, action] = -np.inf  

    return recommended_posts, user_preferences


#Affichage des recommandations avec les cat√©gories des posts et les pr√©f√©rences de l'utilisateur
def display_recommendations(recommended_posts, user_preferences):
    """
    Affiche les posts recommand√©s pour un utilisateur avec ses pr√©f√©rences.

    Parameters:
    recommended_posts (list): Liste des identifiants des posts recommand√©s.
    user_preferences (pandas.Series): Les pr√©f√©rences de l'utilisateur.

    Returns:
    None
    """
    print("------------Posts recommand√©s pour l'utilisateur :------------------")
    for post_id in recommended_posts:
        print(f"Post ID : {post_id}")
        print(f"Cat√©gorie du post : {df['post_categorie_txt'].iloc[post_id]}")
        print("Pr√©f√©rences de l'utilisateur :")
        for i, preference in enumerate(user_preferences):
            print(f"Pr√©f√©rence {i+1} : {preference}")
        print()


#Entra√Ænement
q_table = train_q_learning(df)

#Recommander des posts pour un utilisateur sp√©cifique (user_id)
user_id = 0
recommended_posts, user_preferences = recommend_posts(user_id, q_table)

#Afficher les recommandations
display_recommendations(recommended_posts, user_preferences)


#----------------------------------------METRIQUES-----------------------------------------
#Taux de pertinence
def relevance_rate(recommended_posts, user_preferences):
    num_interesting_posts = sum(post in user_preferences.values for post in recommended_posts)
    return num_interesting_posts / len(recommended_posts)

#Taux de couverture
def coverage_rate(recommended_posts, user_preferences):
    unique_preferences = set(user_preferences.values)
    covered_preferences = [post for post in recommended_posts if post in unique_preferences]
    return len(covered_preferences) / len(unique_preferences) if unique_preferences else 0

# Diversit√© des recommandations
def diversity(recommended_posts):
    unique_categories = set(df['post_categorie_txt'].iloc[recommended_posts])
    return len(unique_categories)

#Popularit√© moyenne des posts recommand√©s
def average_popularity(recommended_posts):
    popularity = df['Likes'].iloc[recommended_posts].mean()
    return popularity

#Nouveaut√© des recommandations
def novelty(recommended_posts, user_id):
    user_likes = df[df['user_id'] == user_id]['post_id'].unique()
    num_new_posts = sum(post not in user_likes for post in recommended_posts)
    return num_new_posts / len(recommended_posts)

#√âvaluer les recommandations
def evaluate_recommendations(recommended_posts, user_preferences, user_id):
    relevance = relevance_rate(recommended_posts, user_preferences)
    coverage = coverage_rate(recommended_posts, user_preferences)
    div = diversity(recommended_posts)
    popularity = average_popularity(recommended_posts)
    nov = novelty(recommended_posts, user_id)
    return relevance, coverage, div, popularity, nov

#M√©triques d'√©valuation
relevance, coverage, div, popularity, nov = evaluate_recommendations(recommended_posts, user_preferences, user_id)

# Afficher les m√©triques
print("------------------M√©triques d'√©valuation des recommandations :----------------")
print(f"Taux de pertinence : {relevance}")
print(f"Taux de couverture : {coverage}")
print(f"Diversit√© des recommandations : {div}")
print(f"Popularit√© moyenne des posts recommand√©s : {popularity}")
print(f"Nouveaut√© des recommandations : {nov}")
```

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    Ces am√©liorations ont conduit √† une augmentation significative de la pertinence et de la couverture des recommandations, tout en maintenant une diversit√© et une popularit√© ad√©quate des posts recommand√©s. Toutefois, la nouveaut√© des recommandations reste un d√©fi, car le mod√®le a tendance √† favoriser des posts d√©j√† connus ou align√©s avec les pr√©f√©rences existantes, ce qui peut limiter l'exposition √† de nouveaux contenus.
</p>

<p style="font-family: 'Arial', sans-serif; font-size: 14px;">
    En conclusion, l'int√©gration du Q-learning avec une approche r√©compense ajust√©e pour les pr√©f√©rences des utilisateurs a permis de d√©velopper un syst√®me de recommandation dynamique et √©volutif, offrant des suggestions de plus en plus cibl√©es et pertinentes aux utilisateurs.
</p>


```python
# Fonction pour d√©terminer si le post correspond aux pr√©f√©rences de l'utilisateur
def is_preferred(post_cat, preferences):
    return post_cat in preferences

# Choisir une action avec une politique Œµ-d√©croissante
def choose_action(state, q_table, exploration_rate, step):
    if np.random.uniform(0, 1) < exploration_rate * (0.99 ** step):  # D√©croissance exponentielle de epsilon
        return np.random.choice(q_table.shape[1])
    else:
        return np.argmax(q_table[state, :])

# Mettre √† jour la table Q avec des r√©compenses contextuelles
def update_q_table(q_table, state, action, reward, next_state, learning_rate, discount_factor, preferences, post_cat):
    # R√©compense suppl√©mentaire si le post est pr√©f√©r√©
    extra_reward = 1 if is_preferred(post_cat, preferences) else 0
    reward += extra_reward

    best_next_action = np.argmax(q_table[next_state, :])
    td_target = reward + discount_factor * q_table[next_state, best_next_action]
    td_error = td_target - q_table[state, action]
    q_table[state, action] += learning_rate * td_error

# Entra√Ænement du mod√®le Q-Learning avec des am√©liorations
def train_q_learning(df, learning_rate=0.1, discount_factor=0.9, exploration_rate=0.1, num_episodes=1000):
    num_users = df['user_id'].nunique()
    num_posts = df['post_id'].nunique()
    q_table = np.zeros((num_users, num_posts))

    for episode in range(num_episodes):
        user_id = np.random.randint(0, num_users)
        user_df = df[df['user_id'] == user_id]
        state = user_id

        for step in range(user_df.shape[0]):
            action = choose_action(state, q_table, exploration_rate, step)
            reward = user_df.iloc[action]['Likes']
            post_cat = user_df.iloc[action]['post_categorie_txt']
            preferences = user_df.iloc[0][['user_preference_1', 'user_preference_2', 'user_preference_3', 'user_preference_4', 'user_preference_5']].values
            next_state = user_id
            update_q_table(q_table, state, action, reward, next_state, learning_rate, discount_factor, preferences, post_cat)

    return q_table

# M√©triques de performance am√©lior√©es
def relevance_rate(recommended_posts, user_preferences):
    recommended_categories = [df['post_categorie_txt'].iloc[post] for post in recommended_posts]
    num_relevant_posts = sum(cat in user_preferences.values for cat in recommended_categories)
    return num_relevant_posts / len(recommended_posts)

def coverage_rate(recommended_posts, user_preferences):
    recommended_categories = set(df['post_categorie_txt'].iloc[post] for post in recommended_posts)
    unique_preferences = set(user_preferences.values)
    covered_preferences = recommended_categories.intersection(unique_preferences)
    return len(covered_preferences) / len(unique_preferences) if unique_preferences else 0

def diversity(recommended_posts):
    unique_categories = set(df['post_categorie_txt'].iloc[post] for post in recommended_posts)
    return len(unique_categories)

def average_popularity(recommended_posts):
    popularity = df['Likes'].iloc[recommended_posts].mean()
    return popularity

def novelty(recommended_posts, user_id):
    user_likes = df[df['user_id'] == user_id]['post_id'].unique()
    num_new_posts = sum(post not in user_likes for post in recommended_posts)
    return num_new_posts / len(recommended_posts)

# √âvaluation des recommandations
def evaluate_recommendations(recommended_posts, user_preferences, user_id):
    relevance = relevance_rate(recommended_posts, user_preferences)
    coverage = coverage_rate(recommended_posts, user_preferences)
    div = diversity(recommended_posts)
    popularity = average_popularity(recommended_posts)
    nov = novelty(recommended_posts, user_id)
    return relevance, coverage, div, popularity, nov

# Afficher les m√©triques
def print_metrics(relevance, coverage, diversity, popularity, novelty):
    print("------------------M√©triques d'√©valuation des recommandations :----------------")
    print(f"Taux de pertinence : {relevance}")
    print(f"Taux de couverture : {coverage}")
    print(f"Diversit√© des recommandations : {diversity}")
    print(f"Popularit√© moyenne des posts recommand√©s : {popularity}")
    print(f"Nouveaut√© des recommandations : {novelty}")

# Exemple d'usage
q_table = train_q_learning(df)
user_id = 0
recommended_posts, user_preferences = recommend_posts(user_id, q_table)
display_recommendations(recommended_posts, user_preferences)
relevance, coverage, div, popularity, nov = evaluate_recommendations(recommended_posts, user_preferences, user_id)
print_metrics(relevance, coverage, div, popularity, nov)
```

    En r√©sum√©, cette deuxi√®me am√©lioration apporte une couche suppl√©mentaire de personnalisation et d'adaptation dans le mod√®le de recommandation, en prenant en compte des facteurs externes tels que l'√¢ge et le pays. Cela vise √† augmenter la pertinence et la satisfaction de l'utilisateur, tout en explorant comment ces facteurs d√©mographiques influencent les pr√©f√©rences et les comportements de contenu.

```python
# Fonction pour d√©terminer si le post correspond aux pr√©f√©rences de l'utilisateur
def is_preferred(post_cat, preferences):
    return post_cat in preferences

# Choisir une action avec une politique Œµ-d√©croissante
def choose_action(state, q_table, exploration_rate, step):
    if np.random.uniform(0, 1) < exploration_rate * (0.99 ** step):  # D√©croissance exponentielle de epsilon
        return np.random.choice(q_table.shape[1])
    else:
        return np.argmax(q_table[state, :])

# Mettre √† jour la table Q avec des r√©compenses contextuelles
def update_q_table(q_table, state, action, reward, next_state, learning_rate, discount_factor, preferences, post_cat, age, country):
    # R√©compense suppl√©mentaire si le post est pr√©f√©r√©
    extra_reward = 1 if is_preferred(post_cat, preferences) else 0
    reward += extra_reward
    
    # Adjustments based on user's age and country
    if age > 30:
        reward += 0.5  # Assuming older users value different content
    if country in ["USA", "UK"]:
        reward += 0.5  # Assuming users from these countries have higher engagement

    best_next_action = np.argmax(q_table[next_state, :])
    td_target = reward + discount_factor * q_table[next_state, best_next_action]
    td_error = td_target - q_table[state, action]
    q_table[state, action] += learning_rate * td_error

# Entra√Ænement du mod√®le Q-Learning avec des am√©liorations
def train_q_learning(df, learning_rate=0.1, discount_factor=0.9, exploration_rate=0.1, num_episodes=1000):
    num_users = df['user_id'].nunique()
    num_posts = df['post_id'].nunique()
    q_table = np.zeros((num_users, num_posts))

    for episode in range(num_episodes):
        user_id = np.random.randint(0, num_users)
        user_df = df[df['user_id'] == user_id]
        state = user_id

        for index, row in user_df.iterrows():
            action = choose_action(state, q_table, exploration_rate, index)
            reward = row['Likes']
            post_cat = row['post_categorie_txt']
            preferences = [row[f'user_preference_{i}'] for i in range(1, 6)]
            age = row['user_age']
            country = row['user_pays_origine']
            next_state = state  # State is static per user
            update_q_table(q_table, state, action, reward, next_state, learning_rate, discount_factor, preferences, post_cat, age, country)

    return q_table

# Recommender system
def recommend_posts(user_id, q_table, num_recommendations=10):
    """
    Recommande des posts √† un utilisateur en utilisant la table Q.

    Parameters:
    user_id (int): L'identifiant de l'utilisateur pour lequel recommander des posts.
    q_table (numpy.array): La table Q contenant les valeurs Q pour chaque paire (√©tat, action).
    num_recommendations (int, optionnel): Le nombre de posts √† recommander. Par d√©faut, 10.

    Returns:
    tuple: Un tuple contenant une liste des identifiants des posts recommand√©s et les pr√©f√©rences de l'utilisateur.
    """
    #S√©lection des interactions de l'utilisateur sp√©cifi√©
    user_df = df[df['user_id'] == user_id]
    #Extraction des pr√©f√©rences de l'utilisateur
    user_preferences = user_df[['user_preference_1', 'user_preference_2', 'user_preference_3', 'user_preference_4', 'user_preference_5']].iloc[0]

    recommended_posts = []
    for _ in range(num_recommendations):
        #S√©lection de l'action (post) avec la plus grande valeur Q pour l'utilisateur donn√©
        action = np.argmax(q_table[user_id, :])
        #Ajout du post recommand√© √† la liste des recommandations
        recommended_posts.append(action)
        #Marquer le post comme d√©j√† recommand√© en affectant une valeur n√©gative √† sa valeur Q
        q_table[user_id, action] = -np.inf  

    return recommended_posts, user_preferences

#Affichage des recommandations avec les cat√©gories des posts et les pr√©f√©rences de l'utilisateur
def display_recommendations(recommended_posts, user_preferences):
    """
    Affiche les posts recommand√©s pour un utilisateur avec ses pr√©f√©rences.

    Parameters:
    recommended_posts (list): Liste des identifiants des posts recommand√©s.
    user_preferences (pandas.Series): Les pr√©f√©rences de l'utilisateur.

    Returns:
    None
    """
    print("------------Posts recommand√©s pour l'utilisateur :------------------")
    for post_id in recommended_posts:
        print(f"Post ID : {post_id}")
        print(f"Cat√©gorie du post : {df['post_categorie_txt'].iloc[post_id]}")
        print("Pr√©f√©rences de l'utilisateur :")
        for i, preference in enumerate(user_preferences):
            print(f"Pr√©f√©rence {i+1} : {preference}")
        print()


#Entra√Ænement
q_table = train_q_learning(df)

#Recommander des posts pour un utilisateur sp√©cifique (user_id)
user_id = 0
recommended_posts, user_preferences = recommend_posts(user_id, q_table)

#Afficher les recommandations
display_recommendations(recommended_posts, user_preferences)

# M√©triques de performance am√©lior√©es
def relevance_rate(recommended_posts, user_preferences):
    recommended_categories = [df['post_categorie_txt'].iloc[post] for post in recommended_posts]
    num_relevant_posts = sum(cat in user_preferences.values for cat in recommended_categories)
    return num_relevant_posts / len(recommended_posts)

def coverage_rate(recommended_posts, user_preferences):
    recommended_categories = set(df['post_categorie_txt'].iloc[post] for post in recommended_posts)
    unique_preferences = set(user_preferences.values)
    covered_preferences = recommended_categories.intersection(unique_preferences)
    return len(covered_preferences) / len(unique_preferences) if unique_preferences else 0

def diversity(recommended_posts):
    unique_categories = set(df['post_categorie_txt'].iloc[post] for post in recommended_posts)
    return len(unique_categories)

def average_popularity(recommended_posts):
    popularity = df['Likes'].iloc[recommended_posts].mean()
    return popularity

def novelty(recommended_posts, user_id):
    user_likes = df[df['user_id'] == user_id]['post_id'].unique()
    num_new_posts = sum(post not in user_likes for post in recommended_posts)
    return num_new_posts / len(recommended_posts)

# √âvaluation des recommandations
def evaluate_recommendations(recommended_posts, user_preferences, user_id):
    relevance = relevance_rate(recommended_posts, user_preferences)
    coverage = coverage_rate(recommended_posts, user_preferences)
    div = diversity(recommended_posts)
    popularity = average_popularity(recommended_posts)
    nov = novelty(recommended_posts, user_id)
    return relevance, coverage, div, popularity, nov

# Afficher les m√©triques
def print_metrics(relevance, coverage, diversity, popularity, novelty):
    print("------------------M√©triques d'√©valuation des recommandations :----------------")
    print(f"Taux de pertinence : {relevance}")
    print(f"Taux de couverture : {coverage}")
    print(f"Diversit√© des recommandations : {diversity}")
    print(f"Popularit√© moyenne des posts recommand√©s : {popularity}")
    print(f"Nouveaut√© des recommandations : {novelty}")

# Exemple d'usage
q_table = train_q_learning(df)
user_id = 0
recommended_posts, user_preferences = recommend_posts(user_id, q_table)
display_recommendations(recommended_posts, user_preferences)
relevance, coverage, div, popularity, nov = evaluate_recommendations(recommended_posts, user_preferences, user_id)
print_metrics(relevance, coverage, div, popularity, nov)
```
